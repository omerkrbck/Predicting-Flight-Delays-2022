```python
Airlines_dummies = pd.get_dummies(dfmodel['Airlines'], prefix='Airlines', drop_first=True)
#OriginCityName_dummies = pd.get_dummies(dfmodel['OriginCityName'], prefix='OriginCityName', drop_first=True)
DestCityName_dummies = pd.get_dummies(dfmodel['DestCityName'], prefix='DestCityName', drop_first=True)
Month_dummines = pd.get_dummies(dfmodel['Month'], prefix='Month', drop_first=True) 
#DayofMonth_dummies = pd.get_dummies(dfmodel['DayofMonth'], prefix='DayofMonth', drop_first=True)
DayOfWeek_dummies = pd.get_dummies(dfmodel['DayOfWeek'], prefix='DayOfWeek', drop_first=True)
```

```python
dfmodel = dfmodel.drop(['Airlines', 'OriginCityName', 'DestCityName', 'Month', 'DayofMonth', 'DayOfWeek', 'ArrDelay'], axis=1)
```

```python
dfmodel = pd.concat([dfmodel, Airlines_dummies, DestCityName_dummies, Month_dummines, DayOfWeek_dummies], axis=1)
```

```python
y = dfmodel['Flight_Status'].copy()
X = dfmodel.drop(['Flight_Status'], axis=1).copy()

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7, random_state=42)

#Validation split
X_train, X_valid, y_train, y_valid = train_test_split(X_train, y_train, train_size=0.7, random_state=42)
```

```python
model = DecisionTreeClassifier(max_depth=5)
model.fit(X_train, y_train)
model_acc = model.score(X_valid, y_valid)
print("Test Accuracy: {:.2f}%".format(model_acc * 100))

y_true = np.array(y_valid)
y_pred = model.predict(X_valid)

cm = confusion_matrix(y_true, y_pred)
clr = classification_report(y_true, y_pred, target_names=["NOT DELAYED", "DELAYED"])

plt.figure(figsize=(8, 8))
sns.heatmap(cm, annot=True, vmin=0, fmt='g', cmap='Blues', cbar=False)
plt.xticks(np.arange(2) + 0.5, ["NOT DELAYED", "DELAYED"])
plt.yticks(np.arange(2) + 0.5, ["NOT DELAYED", "DELAYED"])
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix")
plt.show()

print("Classification Report:\n----------------------\n", clr)
```

# Maximum depth performance

```python
from sklearn.metrics import accuracy_score
max_depth_list = [5,10,15,20,50]

accuracy = []

for x in max_depth_list:
    dtc = DecisionTreeClassifier(max_depth=x) 
    dtc.fit(X_train, y_train)
    accuracy.append(dtc.score(X_valid, y_valid))

plt.plot(max_depth_list, accuracy) 
plt.plot(max_depth_list[0], accuracy[0], marker='o') 
plt.grid() 
plt.ylabel('accuracy') 
plt.xlabel('max depth') 
```

```python
DT_regr = DecisionTreeClassifier(max_depth=10, random_state=420)
DT_model = DT_regr.fit(X_train,y_train)
DT_predictions = pd.DataFrame(DT_regr.predict(X_test))
print("MSE =", metrics.mean_squared_error(DT_predictions, y_test))
DT_accuracy = DT_model.score(X_test, y_test)
print('DT Accuracy on Test Set = '+ str(DT_accuracy.round(2)))
```

```python
y_true = np.array(y_test)
y_pred = DT_model.predict(X_test)
cm = confusion_matrix(y_true, y_pred)
clr = classification_report(y_true, y_pred, target_names=["NOT DELAYED", "DELAYED"])

plt.figure(figsize=(8, 8))
sns.heatmap(cm, annot=True, vmin=0, fmt='g', cmap='Blues', cbar=False)
plt.xticks(np.arange(2) + 0.5, ["NOT DELAYED", "DELAYED"])
plt.yticks(np.arange(2) + 0.5, ["NOT DELAYED", "DELAYED"])
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix")
plt.show()

print("Classification Report:\n----------------------\n", clr)
```
