```python
import pandas as pd
import numpy as np
```

The U.S. Department of Transportation's (DOT) Bureau of Transportation Statistics tracks the on-time performance of domestic flights operated by large air carriers. 
This dataset is collected from the Bureau of Transportation Statistics, Govt. of the USA. This data is open-sourced under U.S. Govt. Works.  I dowload 12 csv file that represents each month of 2022. This dataset contains all US domestic flights in 2022.

Here is the URL of the dataset. 

https://www.transtats.bts.gov/DL_SelectFields.aspx?gnoyr_VQ=FGK&QO_fu146_anzr=b0-gvzr

There are 120 features in each month csv. I prepared a list which shows the general information about flights. 
I choose 19 columns from every month of 2022, and then I concat them and save it as csv file. Dataset has 19 features in average with 7 million rows.
